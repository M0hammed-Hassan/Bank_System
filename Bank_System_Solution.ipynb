```python
class User:
    def __init__(self,__name,__age,__gender):
        self.__name=__name
        self.__age=__age
        self.__gender=__gender
    def user_details(self):
        print(f'''User details:\nUser name:{self.__name}\nUser age:{self.__age}\nUser gender:{self.__gender}''')
    def set_name(self,new_name):
        self.__name=new_name
    def get_name(self):
        return self.__name
    def set_age(self,new_age):
        self.__age=new_age
    def get_age(self):
        return self.__age
    def set_gender(self,new_gender):
        self.__gender=new_gender
    def get_gender(self):
        return self.__gender
class Bank(User):
    def __init__(self,__name,__age,__gender):
        super().__init__(__name,__age,__gender)
        self.__balance=0
    def deposits(self,__amount):
        self.__balance=self.__balance+__amount
        print(f'Your balance updated to {self.__balance} EGP')
    def withdraw(self,__amount):
        if self.__balance>=__amount:
            self.__balance=self.__balance-__amount
            print(f'Your balance updated to {self.__balance} EGP')
        else:
            print('Your balance not enough')
    def view_balance(self):
        print(f'Your balance = {self.__balance} EGP')
class Security(Bank):
    def __init__(self,__name,__age,__gender,__ID,__password):
        super().__init__(__name,__age,__gender)
        self.__ID=__ID
        self.__password=__password
    def set_id(self,new_id):
        self.__ID=new_id
    def get_id(self):
        return self.__ID
    def set_pass(self,new_pass):
        self.__password=new_pass
    def get_pass(self):
        return self.__password
```
